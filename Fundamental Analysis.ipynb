### Correlation Analysis

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
stock1='WTI'
stock2='DAL'
df=yf.download([stock1, stock2])
```

```python
df
```

```python
df.head()
```

```python
df.tail()
```

```python
df.dropna(inplace=True)
```

```python
df.plot(subplots=True, figsize=(10,6));
```

```python
df['Adj Close'].plot(subplots=True, figsize=(10, 6));
```

```python
df=df['Adj Close']
```

```python
df.loc['2010':'2024'].plot(secondary_y=stock2, figsize=(10,6));
```

```python
df.index
```

```python
df_sm=df.loc['2010':'2024']
```

```python
rets=np.log(df_sm/df_sm.shift(1))
```

```python
rets.dropna(inplace=True)
```

```python
rets.plot(subplots=True, figsize=(10,6))
```

```python
pd.plotting.scatter_matrix(rets,
                           alpha=0.2,
                           diagonal='hist',
                           hist_kwds={'bins':35},
                           figsize=(10,6))
```

### OLS Regression

```python
reg=np.polyfit(rets[stock1], rets[stock2], deg=1)
ax=rets.plot(kind='scatter', x=stock1, y=stock2, figsize=(10,6))
ax.plot(rets[stock1], np.polyval(reg, rets[stock1]), 'r', lw=2)
```

### Correlation

```python
rets.corr()
```

```python
ax=rets[stock1].rolling(window=252).corr(rets[stock2]).plot(figsize=(10,6))
ax.axhline(rets.corr().iloc[0,1], c='r')                                        
```

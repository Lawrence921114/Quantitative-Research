```python
import pandas as pd

# 1. 先確保有 X+1 欄位
df['增資公告日(X+1)'] = df['增資公告日(X)'] + pd.Timedelta(days=1)

# 2. 合併取得 X 收盤價
df_merged = df.merge(
    merged_df[['證券代碼','年月日','收盤價(元)']], 
    left_on=['證券代碼','增資公告日(X)'],
    right_on=['證券代碼','年月日'],
    how='left'
)
df_merged['X 收盤價'] = df_merged['X 收盤價'].fillna(df_merged['收盤價(元)'])
df_merged.drop(columns=['年月日', '收盤價(元)'], inplace=True)

# 3. 合併取得 X+1「開盤價」與「收盤價」
df_merged = df_merged.merge(
    # 一次帶入「開盤價(元)」、「收盤價(元)」，並改成「X+1 開盤價」、「X+1 收盤價」
    merged_df[['證券代碼','年月日','開盤價(元)','收盤價(元)']].rename(
        columns={
            '開盤價(元)': 'X+1 開盤價',
            '收盤價(元)': 'X+1 收盤價'
        }
    ),
    left_on=['證券代碼','增資公告日(X+1)'],
    right_on=['證券代碼','年月日'],
    how='left'
)
df_merged.drop(columns=['年月日'], inplace=True)

# 4. 檢視結果
df_merged.head()
```

```python
# 1. 用 _y 欄位的值填補 _x 欄位中的 NaN
df_merged['X+1 開盤價_x'] = df_merged['X+1 開盤價_x'].fillna(df_merged['X+1 開盤價_y'])
df_merged['X+1 收盤價_x'] = df_merged['X+1 收盤價_x'].fillna(df_merged['X+1 收盤價_y'])

# 2. 最後可根據需求選擇留下哪個欄位
#    例如把 _x 欄位 rename 成想要的正式欄位名稱
df_merged.rename(
    columns={
        'X+1 開盤價_x': 'X+1 開盤價',
        'X+1 收盤價_x': 'X+1 收盤價'
    },
    inplace=True
)

# 3. 也可以將 _y 欄位直接刪除，避免重複
df_merged.drop(columns=['X+1 開盤價_y','X+1 收盤價_y'], inplace=True, errors='ignore')

# 檢查最終結果
df_merged.head()
```

### 把剩下NaN填上

```python
df_merged.sort_values(by=['證券代碼','增資公告日(X)'], inplace=True)

columns_to_fill = ['X 收盤價', 'X+1 開盤價', 'X+1 收盤價']

# 先後填
df_merged[columns_to_fill] = (
    df_merged
    .groupby('證券代碼')[columns_to_fill]
    .apply(lambda g: g.bfill().ffill())
)

# 若還有 NaN，就用 0(或其他值) 補
df_merged[columns_to_fill] = df_merged[columns_to_fill].fillna(0)
```

```python
df_merged.sort_values(by=['證券代碼','增資公告日(X)'], inplace=True)
columns_to_fill = ['X 收盤價', 'X+1 開盤價', 'X+1 收盤價']

df_merged[columns_to_fill] = (
    df_merged
    .groupby('證券代碼')[columns_to_fill]
    .transform(lambda g: g.bfill().ffill())
)

df_merged[columns_to_fill] = df_merged[columns_to_fill].fillna(0)
```
